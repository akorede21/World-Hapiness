```python
# Importing Another data set that has countries information
country = pd.read_csv('https://raw.githubusercontent.com/lukes/ISO-3166-Countries-with-Regional-Codes/f3d559220372d99a2d5c13a213cf9c2bceb8c6d9/all/all.csv')
country.head()

```

```python
#drop unnecessary columns
country = country.drop(['country-code','alpha-2','alpha-3','iso_3166-2','sub-region','intermediate-region','region-code','sub-region-code','intermediate-region-code'], axis = 1)
```

```python
# The region column contain null value so I will drop it
country.dropna(axis=0,inplace=True)
country.region.unique()
```

```python
# Region name
country.region.unique()
```

```python
#Create a dataframe of each region to fill it in our data.
Asia = country.query("region == 'Asia'")
Europe = country.query("region == 'Europe'")
Africa = country.query("region == 'Africa'")
Oceania = country.query("region == 'Oceania'")
```

```python
# Keep in mind this list is contained both north and south America countries
Americas = country.query("region == 'Americas'")
Americas.name.unique()
```

```python
# This function fill region name based on the argument contained in  the list of countries in each continent.
def fill_region(df,r_set,r_name):
    df[df['Country name'].isin(r_set) == True] = df.replace('-',r_name)
```

```python

world_data['Continent'] = '-'
```

```python
# Create a list of each region
fill_region(world_data,Asia.name,"Asia")
fill_region(world_data,Europe.name,"Europe")
fill_region(world_data,Africa.name,"Africa")
fill_region(world_data,Americas.name,"South America")
fill_region(world_data,Oceania.name,"Oceania")
```

```python
# Checking the Africa Happness According to the Happiness score
africa2005 = africa[africa['Year']==2005]
africa2005_sorted = africa2005.reset_index(drop=True).sort_values(by='Hapiness score', ascending=False).reset_index(drop=True)
africa2005_sorted['Rank'] = [i+1 for i in range(africa2005_sorted.shape[0])]
africa2005_sorted.head()
```
